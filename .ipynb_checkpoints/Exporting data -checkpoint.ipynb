# Wikipedia Database - Income per county 2009-2013

Problem, the data comes from 2009-2013, but good source for reference of counties 


https://en.wikipedia.org/wiki/List_of_United_States_counties_by_per_capita_income

```python
df_3 = pd.read_excel('./Data/Wikipedia_County_income.xlsx')
df_3.columns = df_3.loc[2]
df_3.drop([0,1,2],inplace = True)

withna_2 = df_3[df_3.Rank.isna()].copy()
df_3 = df_3.dropna()

df_3.reset_index(drop = True,inplace = True)
df_3.drop('Rank',axis =1,inplace = True)

for i in df_3.columns[2:]:
    df_3[i] = pd.to_numeric(df_3[i])
    
df_3.head()
```

# Government data income per county 2015-2017

https://www.bea.gov/data/income-saving/personal-income-county-metro-and-other-areas

```python
df_4 = pd.read_excel('./Data/gov_countyincome_2015_17.xlsx')
df_4.columns = ['County','Per cap 2015','Per cap 2016','Per cap 2017','rank_2017','perc change 2016','perc change 2017','rankperc_2017']
df_4.drop([0,1,2,4],inplace = True)

withna = df_4[df_4['rank_2017'].str.contains('--') == True].copy()
df_4.dropna(inplace=True)
df_4.drop(withna.index,inplace = True)

df_4.reset_index(drop = True,inplace= True)
df_4.head()
```

# Realizing that there are some different Counties in the US between 2009 and 2017
